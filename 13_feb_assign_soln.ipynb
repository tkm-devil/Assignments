# Question 6
Answer : Some best practices for exception handling :-
        
        1. Always try to use specific exception and not the super class.
        2. Always print proper and appropriate message.
        3. Always try to log your errors in a log file.
        4. Avoid using multiple exception class on a single block of code, be specific.
        5. Document all your errors this will be helpful in future testing and analysis.
        6. Cleanup all your resources.
