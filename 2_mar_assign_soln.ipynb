# Question 2
Answer : Scatter plot - It is a type of plot that shows the relationship between two variables as a collection of points in 2-D plane. Each point represents the value of the two variables for a single observation, and the position of the point on the plot is determined by the values of the variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.title("Scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Question 3
Answer : The subplot() function in Matplotlib is used to create multiple plots in a single figure. It takes three arguments: the number of rows, the number of columns, and the index of the current plot.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

fig.tight_layout()
```

# Question 4
Asnwer : Bar plot - It is a type of plot that represents categorical data with rectangular bars with lengths proportional to the values they represent.
         Uses - It is used to compare the magnitudes of different quantities or categories. They are commonly used in business, finance, and economics to visualize data such as sales figures, profits, market shares.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Vertical Plot
plt.bar(company, profit)
plt.title("Company profits")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
# Horizontal plot
plt.barh(company, profit)
plt.title("Company profits")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

# Question 5
Answer : Box plot - It is a type of plot that displays the distribution of a dataset. The box represents the interquartile range (IQR), which is the range between the 25th and 75th percentiles of the data. The whiskers represent the range of the data outside the IQR, and the points outside the whiskers represent outliers.

        Uses - It used to identify the skewness and outliers in the data, compare the distributions of different datasets, and visualize the spread and central tendency of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2])
plt.title("Box plot")
plt.xlabel("Box")
plt.ylabel("Value")
plt.show()
```
