# Question 1
Answer : Five common plots in seaborn :-

        1. Scatter Plot: It is used to display the relationship between two continuous variables. It is useful for detecting patterns, clusters, and outliers in data.
        2. Bar Plot: It is used to compare categorical data using rectangular bars with heights or lengths proportional to the values they represent. It is useful for visualizing the distribution, frequency, or proportion of different categories or groups in data. 
        3. Histogram: It is used to display the distribution of a continuous variable by dividing it into bins and counting the frequency or density of observations in each bin. It is useful for examining the shape, center, and spread of data, as well as identifying outliers, skewness.
        4. Line Plot: It is used to display the trend or pattern of a continuous variable over time or space. It is useful for visualizing changes, trends, and patterns in data, as well as comparing multiple variables or groups over time. 
        5. Heatmap: It is used to display the correlation or relationship between two or more variables using colors to represent the intensity or strength of the relationship. It is useful for identifying patterns, trends, and outliers in data, as well as exploring the relationships between different variables.

```python
fmri_Data = sns.load_dataset('fmri')
```

```python
fmri_Data.head()
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_Data)
```
