# Question 5
Answer : Distplot - It is a visualization function in seaborn that is used to plot a univariate distribution of observations. It combines a histogram and a kernel density estimate plot. It is useful for visualizing the distribution of a dataset and for identifying outliers.

```python
import numpy as np
```

```python
np.random.seed(123)
data = np.random.normal(0, 1, 1000)
```

```python
distplot = px.histogram(data, nbins=80, marginal='violin', opacity=0.7,barmode='overlay')
distplot.update_traces(marker=dict(line=dict(width=1, color='Black')))
distplot.show()
```
