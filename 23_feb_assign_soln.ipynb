        Dataframe - A DataFrame is a 2-dimensional labeled data structure with columns of potentially different data types. A DataFrame can be created by passing a dictionary of arrays, lists, or Series, or by reading data from a file.
        Series - A Series is a one-dimensional labeled array that can hold any data type (integers, strings, floating-point numbers, Python objects, etc.). A Series is similar to a column in a spreadsheet or a SQL table, and it has a label (called an index) for each element in the array.
