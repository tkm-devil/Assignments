```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

# Question 3
Answer : App routing is a fundamental concept in Flask that refers to the mechanism of mapping URL paths to the functions that handle them. In other words, it allows developers to specify what happens when a user visits a particular URL path in their application.
        We use app routes to define the behavior of our application for different URL paths.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def display():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route('/')
def details():
    return 'Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999'

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

# Question 5
Answer : The 'url_for()' function is used in Flask to dynamically generate URLs for a given view function.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def profile(username):
    return f'Welcome, {username}!'

@app.route('/')
def index():
    user_url = url_for('profile', username='Tilak')
    return f"The URL for Tilak's profile is {user_url}"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
