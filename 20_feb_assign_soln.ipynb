        GET - The GET method is used to request a resource or retrieve data from the server. When a client sends a GET request, the request parameters are sent in the URL. This means that the data being transmitted is visible in the URL, and there is a limit to how much data can be sent.
        
        POST - The POST method, is used to submit data to the server. When a client sends a POST request, the request parameters are sent in the body of the request. This means that the data being transmitted is not visible in the URL, and there is no limit to how much data can be sent.

Answer : Templates are used to generate dynamic HTML pages that are returned to the user in response to a request. Templates are HTML files that contain placeholders for dynamic content, such as variables, loops, and conditional statements.

        render_template() - It is used to render a template and return the resulting HTML page to the user. This function takes the name of the template file as its first argument and any data that should be passed to the template as keyword arguments. The function then looks for the template file in a directory called templates in the root directory of the Flask application.
